```python
import pandas as pd

# Load the datasets
file_path_working = 'goal_ticks_working.csv'
file_path_not_working = 'goal_ticks_not_working.csv'
file_path_new = 'goal_ticks_new_kartID.csv'

# Read the first few lines of each file to check the structure
df_working = pd.read_csv(file_path_working, header=None)
df_not_working = pd.read_csv(file_path_not_working, header=None)
df_new = pd.read_csv(file_path_new, header=None)

print("Working dataset structure:")
print(df_working.head())
print(f"Number of columns: {df_working.shape[1]}\n")

print("Not working dataset structure:")
print(df_not_working.head())
print(f"Number of columns: {df_not_working.shape[1]}\n")

print("New dataset structure:")
print(df_new.head())
print(f"Number of columns: {df_new.shape[1]}\n")

# Compare with the expected number of columns
expected_columns = 19  # The number of features plus the target column

if df_working.shape[1] != expected_columns:
    print(f"Error: Working dataset has {df_working.shape[1]} columns but expected {expected_columns}.")
if df_not_working.shape[1] != expected_columns:
    print(f"Error: Not working dataset has {df_not_working.shape[1]} columns but expected {expected_columns}.")
if df_new.shape[1] != expected_columns:
    print(f"Error: New dataset has {df_new.shape[1]} columns but expected {expected_columns}.")
```
